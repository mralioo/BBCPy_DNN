```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable, ImageGrid, inset_locator
import mne
```

```python
train_data = smr_datamodule.train_data
eeg_data = train_data.data
# Create an events array from train_data.y and event times
events = []
for i, label in enumerate(train_data.y):
    events.append([i, 0, label])

# Convert the events list to a NumPy array
events = np.array(events)

info = mne.create_info(ch_names=train_data.chans, sfreq=train_data.fs, ch_types=['eeg'] * train_data.nCh)
# Create a RawArray object
# Create an EpochsArray object
epochs = mne.EpochsArray(eeg_data, info, events=events, event_id = {'Left': 0, 'Right': 1, 'Up':2, 'Down':3})
```

```python
L_epochs = epochs["Left"]
R_epochs = epochs["Right"]
U_epochs = epochs["Up"]
D_epochs = epochs["Down"]

L_channels = ['C3', 'CP3']
R_channels = ['C4','C2','C6','FC4','CP4']
```

```python
import matplotlib.pyplot as plt
from mne import create_info
import numpy as np

# Assuming you have your EEG data in epochs_data (EpochsArray)
# Define a custom color for the picked channels
channel_colors = {'C3': 'blue', 'C4': 'red'}

# Compute the PSD
psds = R_epochs.compute_psd(picks=["C3", "C4"], fmin=0.1, fmax=50., n_jobs=1)
freqs= psds.freqs
psd = psds.get_data()


# Compute the average PSD over epochs
avg_psd_C3 = np.mean(10 * np.log10(psd[:, 0]), axis=0)
avg_psd_C4 = np.mean(10 * np.log10(psd[:, 1]), axis=0)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot PSD for C3
ax.plot(freqs, avg_psd_C3, color=channel_colors['C3'], label='C3')

# Plot PSD for C4
ax.plot(freqs, avg_psd_C4, color=channel_colors['C4'], label='C4')

# Add a legend
ax.legend()


# Set labels and title
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Power/Frequency (dB/Hz)')
ax.set_title('Right Event : Power Spectral Density')

# Show the plot
plt.show()
```
